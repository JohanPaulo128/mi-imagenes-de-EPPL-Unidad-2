 Redacción del Problema
En el estudio de la geometría euclidiana, las figuras circulares y esféricas son fundamentales para comprender cómo medir superficies en dos dimensiones y espacios en tres dimensiones. Un círculo se define como el conjunto de puntos que se encuentran a una misma distancia (radio) de un punto central. A partir de esta figura, podemos determinar dos magnitudes esenciales: el área, que representa la superficie total encerrada dentro del círculo, y el perímetro (o circunferencia), que mide la longitud de su borde exterior.
Por otro lado, la esfera es la extensión tridimensional del círculo. Su volumen indica la cantidad de espacio físico que ocupa el sólido en el espacio. Para resolver problemas prácticos de manera eficiente, es necesario aplicar fórmulas matemáticas precisas que dependen únicamente del radio de la figura. El siguiente programa interactivo en Python permite calcular estas magnitudes de forma sencilla, utilizando un menú visual y emojis para mejorar la experiencia del usuario.

1.   Fórmulas a Utilizar
2.   Área del círculo: A=π⋅r2A=π⋅r 2
3.   Perímetro del círculo: P=2⋅π⋅rP=2⋅π⋅r
4.   Volumen de la esfera: V=43⋅π⋅r3V= 34​ ⋅π⋅r 3
(Donde π≈3.14159...π≈3.14159... y rr es el radio)

```python
# Programa realizado por Johan Paulo Aros Reveles NL = 4
# Programa de calculo geometrico
# Formulas utilizadas
# A=π⋅r 2
# P=2⋅π⋅r
#V= 34​ ⋅π⋅r 3

import math

# ═══════════════════════════════════════════════════════════════
# FUNCIÓN: areacirculo()
# Calcula el área de un círculo usando la fórmula: A = π * r²
# ═══════════════════════════════════════════════════════════════
def areacirculo():
    print("\n" + "🔵" * 40)
    print("📐 CÁLCULO DEL ÁREA DEL CÍRCULO")
    print("🔵" * 40)
    print("Fórmula: Área = π × r²")

    radio = float(input("\n✏️  Introduce el radio del círculo: "))

    area = math.pi * radio ** 2

    print("\n" + "✅" * 40)
    print(f"📊 Radio ingresado: {radio}")
    print(f"🎯 Área del círculo: {area:.4f}")
    print(f"🎯 Área del círculo (redondeada): {round(area, 2)}")
    print("✅" * 40)

# ═══════════════════════════════════════════════════════════════
# FUNCIÓN: perimetrocirculo()
# Calcula el perímetro (circunferencia) usando: P = 2 * π * r
# ═══════════════════════════════════════════════════════════════
def perimetrocirculo():
    print("\n" + "🟢" * 40)
    print("📏 CÁLCULO DEL PERÍMETRO DEL CÍRCULO")
    print("🟢" * 40)
    print("Fórmula: Perímetro = 2 × π × r")

    radio = float(input("\n✏️  Introduce el radio del círculo: "))

    perimetro = 2 * math.pi * radio

    print("\n" + "✅" * 40)
    print(f"📊 Radio ingresado: {radio}")
    print(f"🎯 Perímetro del círculo: {perimetro:.4f}")
    print(f"🎯 Perímetro del círculo (redondeado): {round(perimetro, 2)}")
    print("✅" * 40)

# ═══════════════════════════════════════════════════════════════
# FUNCIÓN: volumenesfera()
# Calcula el volumen de una esfera usando: V = (4/3) * π * r³
# ═══════════════════════════════════════════════════════════════
def volumenesfera():
    print("\n" + "🔴" * 40)
    print("🌍 CÁLCULO DEL VOLUMEN DE LA ESFERA")
    print("🔴" * 40)
    print("Fórmula: Volumen = (4/3) × π × r³")

    radio = float(input("\n✏️  Introduce el radio de la esfera: "))

    volumen = (4/3) * math.pi * radio ** 3

    print("\n" + "✅" * 40)
    print(f"📊 Radio ingresado: {radio}")
    print(f"🎯 Volumen de la esfera: {volumen:.4f}")
    print(f"🎯 Volumen de la esfera (redondeado): {round(volumen, 2)}")
    print("✅" * 40)

# ═══════════════════════════════════════════════════════════════
# FUNCIÓN: menu()
# Muestra el menú principal y controla la navegación del programa
# Usa if, elif, else y while para la interacción
# ═══════════════════════════════════════════════════════════════
def menu():
    opcion = 0

    while opcion != 4:
        print("\n")
        print("╔" + "═" * 50 + "╗")
        print("║" + " " * 10 + "🎓 MENÚ PRINCIPAL 🎓" + " " * 15 + "║")
        print("╠" + "═" * 50 + "╣")
        print("║  1️⃣  Calcular área del círculo            🔵        ║")
        print("║  2️⃣  Calcular perímetro del círculo       🟢        ║")
        print("║  3️⃣  Calcular volumen de la esfera        🔴        ║")
        print("║  4️⃣  Salir del programa                   👋        ║")
        print("╚" + "═" * 50 + "╝")

        opcion = int(input("\n💡 Selecciona una opción (1-4): "))

        if opcion == 1:
            areacirculo()
        elif opcion == 2:
            perimetrocirculo()
        elif opcion == 3:
            volumenesfera()
        elif opcion == 4:
            print("\n" + "🌟" * 50)
            print("║" + " " * 12 + "¡Gracias por usar el programa! 👋" + " " * 6 + "║")
            print("║" + " " * 14 + "¡Hasta luego! 🎓✨" + " " * 14 + "║")
            print("🌟" * 50)
        else:
            print("\n⚠️  Opción no válida. Por favor selecciona 1, 2, 3 o 4.")
            print("💭 Intenta de nuevo...")

# ═══════════════════════════════════════════════════════════════
# PROGRAMA PRINCIPAL
# Llamada a la función menú para iniciar el programa
# ═══════════════════════════════════════════════════════════════

print("\n" + "🎉" * 50)
print("║" + " " * 8 + "BIENVENIDO AL PROGRAMA DE GEOMETRÍA" + " " * 7 + "║")
print("║" + " " * 12 + "Círculos y Esferas 📐🔵🔴" + " " * 11 + "║")
print("🎉" * 50)

menu()
```
